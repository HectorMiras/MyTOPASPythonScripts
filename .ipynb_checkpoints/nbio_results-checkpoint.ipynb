```python
# IRTGValue_analysis.ipynb ────────────────────────────────────────────────
import pathlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ------------------------------------------------------------------------
# 1. Locate files
# ------------------------------------------------------------------------
root      = pathlib.Path('../TOPAS_CellsNPs/work/topas-med0-smallcell0-var0/run1')
phsp_file = root / 'IRTGValue.phsp'
hdr_file  = root / 'IRTGValue.header'     # not used here but handy for units

# ------------------------------------------------------------------------
# 2. Column definitions (per header)
#    1: GValue       (molecules / 100 eV)
#    2: GValue_err   (1 σ)
#    3: Time_ps      (picoseconds)
#    4: Molecule     (string)
# ------------------------------------------------------------------------
cols = ['GValue', 'GValue_err', 'Time_ps', 'Molecule']

df = pd.read_csv(
    phsp_file,
    comment='#',         # ignore lines starting with '#'
    sep=r'\s+',          # whitespace-delimited
    names=cols,
    header=None,
    engine='python'
)

# ── NEW: drop species whose total production is zero ─────────────────────
active_species = (
    df.groupby('Molecule', sort=False)['GValue']
      .sum()                 # total across all times and histories
      .loc[lambda s: s > 0]  # keep only > 0
      .index
)

df = df[df['Molecule'].isin(active_species)].copy()
# ------------------------------------------------------------------------

# ------------------------------------------------------------------------
# 3. Pivot tables: GValue (and its error) vs. time and molecule
# ------------------------------------------------------------------------
pivot      = df.pivot_table(index='Time_ps', columns='Molecule',
                            values='GValue',     aggfunc='mean')
pivot_err  = df.pivot_table(index='Time_ps', columns='Molecule',
                            values='GValue_err', aggfunc='mean')

# Final-time summary
last_time   = pivot.index.max()
summary     = pivot.loc[last_time].to_frame(name='GValue_final')
summary_err = pivot_err.loc[last_time]

# Show interactive summary table
display(summary.style.format('{:.2f}'))

# ------------------------------------------------------------------------
# 4. Final-time bar chart with 1-σ error bars
# ------------------------------------------------------------------------
plt.figure(figsize=(6,4))
plt.barh(
    summary.sort_values('GValue_final').index,
    summary.sort_values('GValue_final')['GValue_final'].values,
    xerr=summary_err.loc[summary.sort_values('GValue_final').index].values,
    align='center',
    ecolor='k',
    capsize=3
)

plt.xlabel(f'GValue (molecules / 100 eV) at t = {last_time:.1f} ps')
plt.title('Final chemical production per species')
plt.tight_layout()
plt.show()

# ------------------------------------------------------------------------
# 5. Time evolution with ±1 σ bands (log-scaled time axis)
# ------------------------------------------------------------------------
plt.figure(figsize=(8,5))
for mol in pivot.columns:
    plt.plot(pivot.index, pivot[mol], label=mol)
    plt.fill_between(
        pivot.index,
        pivot[mol] - pivot_err[mol],
        pivot[mol] + pivot_err[mol],
        alpha=0.2
    )

plt.xscale('log')
plt.xlabel('Time (ps)')
plt.ylabel('GValue (molecules / 100 eV)')
plt.title('Temporal evolution of chemical species')
plt.legend(fontsize='small')
plt.tight_layout()
plt.show()

# ------------------------------------------------------------------------
# 6. Save CSVs for further processing
# ------------------------------------------------------------------------
summary.to_csv(root / 'IRTGValue_GValueFinal.csv')
pivot.to_csv(root / 'IRTGValue_GValue_vs_time.csv')
print('✔  Summary and time-evolution CSVs saved.')

```

```python
# ── IRTG_enhancement_analysis.ipynb ─────────────────────────────────────────
import pathlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# --------------------------------------------------------------------------
# 1. Localiza los ficheros de cada simulación
# --------------------------------------------------------------------------
root_base = pathlib.Path('../TOPAS_CellsNPs/work/topas-med0-smallcell0-var0/run1') 
root_np = pathlib.Path('../TOPAS_CellsNPs/work/topas-med1-smallcell1-var0/run1') 
# ------- baseline (sin AuNP) ---------------------------------------------
phsp_base     = root_base / 'IRTGValue.phsp'        # o el nombre que uses
# ------- con nanopartículas (AuNP) ----------------------------------------
phsp_np       = root_np / 'IRTGValue.phsp'   # cámbialo si es diferente

# Si los encabezados siguen siendo idénticos, no hace falta leerlos.
cols = ['GValue', 'GValue_err', 'Time_ps', 'Molecule']   # mismo formato que antes

def load_irtg(path):
    return pd.read_csv(
        path, comment='#', sep=r'\s+', names=cols, header=None, engine='python'
    )

df_base = load_irtg(phsp_base)
df_np   = load_irtg(phsp_np)

# -------------------------------------------------------------------------
# 2. Pivot tables: rows = time, cols = molecule
# -------------------------------------------------------------------------
pv_base      = df_base.pivot_table(index='Time_ps', columns='Molecule',
                                   values='GValue',      aggfunc='mean').fillna(0)
pv_base_err  = df_base.pivot_table(index='Time_ps', columns='Molecule',
                                   values='GValue_err',  aggfunc='mean').fillna(0)

pv_np        = df_np  .pivot_table(index='Time_ps', columns='Molecule',
                                   values='GValue',      aggfunc='mean').fillna(0)
pv_np_err    = df_np  .pivot_table(index='Time_ps', columns='Molecule',
                                   values='GValue_err',  aggfunc='mean').fillna(0)

# -------------------------------------------------------------------------
# 3. Enhancement factor at common final time
# -------------------------------------------------------------------------
t_final = min(pv_base.index.max(), pv_np.index.max())

G_base  = pv_base.loc[t_final]
G_np    = pv_np  .loc[t_final]
σ_base  = pv_base_err.loc[t_final]
σ_np    = pv_np_err  .loc[t_final]

ratio         = G_np / G_base.replace(0, np.nan)          # avoid /0 warnings
ratio_err     = ratio * np.sqrt((σ_np/G_np)**2 + (σ_base/G_base)**2)

# -------------------------------------------------------------------------
# 4. Remove species with zero production in either simulation
# -------------------------------------------------------------------------
valid = (G_np > 0) & (G_base > 0)
ratio      = ratio[valid]
ratio_err  = ratio_err[valid]

# -------------------------------------------------------------------------
# 5. Bar chart with error bars
# -------------------------------------------------------------------------
ratio_sorted  = ratio.sort_values()
err_sorted    = ratio_err[ratio_sorted.index]

plt.figure(figsize=(7,4))
plt.barh(ratio_sorted.index, ratio_sorted.values,
         xerr=err_sorted.values, align='center', ecolor='k', capsize=3)

plt.axvline(1, ls='--', lw=1, color='grey')
plt.xlabel('Chemical-species enhancement factor (AuNP / baseline)')
plt.title(f'Final enhancement (t = {t_final:.1f} ps)')
plt.tight_layout()
plt.show()

# -------------------------------------------------------------------------
# 6. Save CSV with ratios and errors
# -------------------------------------------------------------------------
out = pd.DataFrame({
    'Enhancement': ratio_sorted,
    'Enhancement_err': err_sorted
})
#out.to_csv('Enhancement_final_with_errors.csv')
print('✔  CSV “Enhancement_final_with_errors.csv” written.')
```

```python
# ── IRTGValue_multi-run_aggregation.ipynb ────────────────────────────────
import pathlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

# ------------------------------------------------------------------------
# 1. Settings
# ------------------------------------------------------------------------
base_dir = pathlib.Path('../TOPAS_CellsNPs/work/topas-med1-smallcell1-var0')
run_template = 'run{}'          # run1, run2, … run20
n_runs = 20
phsp_name = 'IRTGValue.phsp'

out_dir = base_dir / 'results'
out_dir.mkdir(exist_ok=True)

cols = ['GValue', 'GValue_err', 'Time_ps', 'Molecule']

# ------------------------------------------------------------------------
# 2. Load every run into a single MultiIndex DataFrame
# ------------------------------------------------------------------------
all_runs = []
for r in range(1, n_runs + 1):
    phsp_path = base_dir / run_template.format(r) / phsp_name
    if not phsp_path.is_file():
        raise FileNotFoundError(f'{phsp_path} not found')
    df = pd.read_csv(phsp_path, comment='#', sep=r'\s+',
                     names=cols, header=None, engine='python')
    df['Run'] = r
    all_runs.append(df)

df_all = pd.concat(all_runs, ignore_index=True)

# ------------------------------------------------------------------------
# 3. Aggregate time–molecule matrices (mean and SD across runs)
# ------------------------------------------------------------------------
pivot_mean = (df_all
              .groupby(['Time_ps', 'Molecule'])['GValue']
              .mean()
              .unstack('Molecule')
              .sort_index())

pivot_sd   = (df_all
              .groupby(['Time_ps', 'Molecule'])['GValue']
              .std(ddof=1)
              .unstack('Molecule')
              .reindex_like(pivot_mean))

# ------------------------------------------------------------------------
# 4. Final-time summary (mean ± SD over runs)
# ------------------------------------------------------------------------
# Collect the final line of each run, then aggregate
final_rows = []
for r in range(1, n_runs + 1):
    df_r = df_all[df_all['Run'] == r]
    last_t = df_r['Time_ps'].max()
    final_rows.append(df_r[df_r['Time_ps'] == last_t])

df_final = pd.concat(final_rows)        #  rows = n_runs × n_species

summary_mean = (df_final
                .groupby('Molecule')['GValue']
                .mean()
                .rename('Mean_GValue'))

summary_sd   = (df_final
                .groupby('Molecule')['GValue']
                .std(ddof=1)
                .rename('SD_GValue'))

# Remove zero-production species
nonzero = summary_mean > 0
summary_mean = summary_mean[nonzero]
summary_sd   = summary_sd[nonzero]

summary_df = pd.concat([summary_mean, summary_sd], axis=1)

# ------------------------------------------------------------------------
# 5. Final-time bar chart with error bars
# ------------------------------------------------------------------------
plt.figure(figsize=(7,4))
summary_mean_sorted = summary_mean.sort_values()
plt.barh(summary_mean_sorted.index,
         summary_mean_sorted.values,
         xerr=summary_sd.loc[summary_mean_sorted.index].values,
         align='center', ecolor='k', capsize=3)

plt.xlabel('Mean GValue (molecules / 100 eV)')
plt.title('Final chemical production per species (20 runs)')
plt.tight_layout()
plt.show()

# ------------------------------------------------------------------------
# 6. Save aggregates to results7
# ------------------------------------------------------------------------
pivot_mean.to_csv(out_dir / 'IRTGValue_GValue_vs_time_mean.csv')
pivot_sd  .to_csv(out_dir / 'IRTGValue_GValue_vs_time_sd.csv')
summary_df.to_csv(out_dir / 'IRTGValue_GValue_final_mean_sd.csv')

print(f'✔  Aggregated files written to {out_dir}')

# Re-use df_final from the previous cell (one row per run×species at t_final)
# Pivot → index = Run, columns = Molecule, values = GValue
pivot_final = (df_final
               .pivot(index='Run', columns='Molecule', values='GValue')
               .reindex(columns=summary_mean.index))   # keep only non-zero species & same order

# Save per-run table
pivot_final.to_csv(out_dir / 'IRTGValue_GValue_final_per_run.csv')

# --- Violin plot ---------------------------------------------------------
species = pivot_final.columns
data    = [pivot_final[col].values for col in species]

plt.figure(figsize=(max(6, 0.6*len(species)), 4))
plt.violinplot(data, showmeans=True, showmedians=False, showextrema=False)

plt.xticks(np.arange(1, len(species)+1), species, rotation=45, ha='right')
plt.ylabel('Final GValue (molecules / 100 eV)')
plt.title('Distribution of final chemical production across runs')
plt.tight_layout()
plt.show()

print('✔  Per-run final-time table and violin plot generated.')
```
